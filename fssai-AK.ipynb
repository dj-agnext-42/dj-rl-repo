```python


import re
import numpy as np

# Mapping superscript characters to normal digits
superscript_map = str.maketrans("⁰¹²³⁴⁵⁶⁷⁸⁹", "0123456789")

def parse_result(value):
    val_str = str(value).strip().lower()
    val_str = val_str.translate(superscript_map)  # Convert superscript numbers

    # Normalize formats like "4.7x x10(5)" → "4.7 x 10^5"
    val_str = re.sub(r"\s*x\s*x\s*", " x ", val_str)  # Fix double "x"
    val_str = re.sub(r"10\((\d+)\)", r"10^\1", val_str)  # Convert 10(5) → 10^5

    # First, check for normal multiplication (e.g., "1.88 x 1000")
    match_multiply = re.search(r"([\d\.]+)\s*[xX×*]\s*(10+)$", val_str)
    if match_multiply:
        base = float(match_multiply.group(1))
        multiplier = int(match_multiply.group(2))
        return np.multiply(base, multiplier)

    # Check for scientific notation (X x 10^Y OR X x 10Y)
    match_sci = re.search(r"([\d\.]+)\s*[xX×*]\s*10\^?(\d+)", val_str)
    if match_sci:
        base = float(match_sci.group(1))
        exponent = int(match_sci.group(2))
        return base * (10 ** exponent)

def convert_result(report_value):
    report_value = str(report_value).strip().replace('\n', '')

    # Keywords mapping to 0
    zero_keywords = ['absent', 'negative', 'nil', 'conform/s', 'compile/s', 'complies', 
                     'not detected', '-absent', 'conforms', 'free', 'not observed', 'freee from']
    if any(kw in report_value.lower() for kw in zero_keywords) or \
       report_value.lower().startswith('free from') or \
       'it is free from' in report_value.lower():
        return 0.0

    # Keywords mapping to 1
    if report_value.lower() in ['present', 'positive'] or report_value.lower().startswith('present'):
        return 1.0

    # Handle BLQ cases
    if 'BL' in report_value or 'B.L.Q' in report_value:
        pattern = re.compile(r'(?:LOQ:?\s*|<|DL:|of\s*|\s*)(\d+\.?\d*)')
        match = pattern.search(report_value)
        if match:
            return float(match.group(1))
        return 0.0001  # Default for BLQ without specific value

    # Handle ND (Not Detected) and BDL (Below Detection Limit)
    if report_value.startswith('ND') or report_value.startswith('BDL'):
        match = re.search(r'DL[: ]*\s*(\d+\.?\d*)', report_value)
        if match:
            return float(match.group(1))

    if 'bql' in report_value.lower():
        return 0.0001

    # Handle "Less Than X" cases
    match_less_than = re.search(r'less\s*than\s*([\d\.]+)', report_value.lower())
    if match_less_than:
        return float(match_less_than.group(1))

    # Handle "<100 m" or similar cases
    match_m = re.search(r'<\s*(\d+)\s*m?', report_value)
    if match_m:
        return float(match_m.group(1))

    # Use parse_result function for scientific notation & multiplication cases
    parsed_value = parse_result(report_value)
    if parsed_value is not None:
        return parsed_value

    # Handle percentages
    if '%' in report_value:
        try:
            return float(report_value.replace('%', ''))
        except ValueError:
            return None

    # Handle values like "1.085, ug/kg"
    report_value = report_value.replace(',', '')  # Remove commas in numbers

    # Handle cases with empty brackets like "()"
    if '()' in report_value:
        report_value = report_value.replace('()', '')

    # Handle values ending with "*"
    if report_value.endswith('*'):
        report_value = report_value.replace('*', '')

    # Try converting directly to float
    try:
        return float(report_value)
    except ValueError:
        # Extract first numeric value from text
        match = re.search(r'([-+]?\d*\.?\d+)', report_value)
        if match:
            return float(match.group(1))
        return None  # Return None for non-numeric values
```

```python
# Read both sheets from the Excel file
import pandas as pd
import re

df_sheet1 = pd.read_excel('b.xlsx')
df_master = df_new_Master

# Identify the relevant parameter columns based on the given range
start_col = 'Acetamiprid, mg/kg'
end_col = 'FSSR Limits Copper Oxychloride (Copper determined as elemental copper), mg/ kg'

# Get the list of columns that fall within the specified range
param_columns = list(df_sheet1.columns)
start_index = param_columns.index(start_col)
end_index = param_columns.index(end_col)
selected_params = param_columns[start_index:end_index+1]  # Ensure inclusive range

# Extract only parameter names (exclude 'FSSR Limits' ones)
params = [col for col in selected_params if not col.startswith('FSSR Limits')]

# Function to get the appropriate value from Master sheet
def get_master_value(commodity, parameter, variant):
    master_row = df_master[(df_master['Commodity'] == commodity) & 
                          (df_master['Parameter'] == parameter)]
    
    if not master_row.empty:
        if variant in ['Loose', 'Normal']:
            value = master_row['Max FSSAI Acceptable Value'].iloc[0]
        elif variant == 'Organic':
            value = master_row['Max FSSAI Acceptable Value - Organic'].iloc[0]
        else:
            return None
        
        if pd.isna(value):
            return None
        return str(value)
    return None

# Iterate through each row in Sheet1
for idx, row in df_sheet1.iterrows():
    commodity = row['Commodity']
    variant = row['Variant']
    
    for param in params:
        fssr_col = f'FSSR Limits {param}'
        compliance_col = f'{param} Compliance'
        
        if compliance_col not in df_sheet1.columns:
            df_sheet1[compliance_col] = None

        current_limit = df_sheet1.at[idx, fssr_col] if fssr_col in df_sheet1.columns else None
        current_original_value=df_sheet1.at[idx, param] if param in df_sheet1.columns else None
        current_value = convert_result(current_original_value)
        
        # Always get the master value for FSSR Limits
        master_value = get_master_value(commodity, param, variant)
        # print(master_value)

        if master_value is not None:
            df_sheet1.at[idx, fssr_col] = f"{current_limit} {master_value}"

        # If the current_value is empty, skip compliance check but keep FSSR Limits updated
        if pd.isna(current_value) or current_value == '':
            df_sheet1.at[idx, compliance_col] = ''
            continue

        # Extract numeric limit value
        # Extract numeric limit value safely
        try:
            numeric_values = re.findall(r'\d+\.?\d*', str(master_value))  # Find all numeric values
            limit_value = float(numeric_values[0]) if numeric_values else None  # Check if list is empty
        except (ValueError, IndexError):
            limit_value = None


        if limit_value is not None:
            decimal_places = len(str(limit_value).split('.')[-1]) if '.' in str(limit_value) else 0
            rounded_value = round(float(current_value), decimal_places)
            df_sheet1.at[idx, compliance_col] = 'Safe' if rounded_value <= limit_value else 'Unsafe'


# Save the updated DataFrame
df_sheet1.to_excel('c.xlsx', index=False)
print("Successfully updated all FSSR Limits columns and added compliance checks")
```

```python
import pandas as pd
import ast

# Load the result dataframe containing 'Commodity', 'Parameter_banned', 'Parameter_not_banned'
result_df = pd.read_excel("differ_2.xlsx")

# Load the final Excel file
file_path = "final_final_2.xlsx"  # Update this with the actual file path
output_file = "reordered_final.xlsx"

xls = pd.ExcelFile(file_path)
sheet_dict = {}

for sheet_name in xls.sheet_names:
    if sheet_name == "Original data":  # Skip the "Original data" sheet
        continue

    df = xls.parse(sheet_name)

    # Get the relevant row in 'result_df' matching the sheet name (commodity)
    commodity_row = result_df[result_df["Commodity"] == sheet_name]

    if not commodity_row.empty:
        banned_parameters = commodity_row["Parameter_banned"].values[0]
        not_banned_parameters = commodity_row["Parameter_not_banned"].values[0]
    
        # Convert from string representation of list to actual list (if needed)
        if isinstance(banned_parameters, str):
            banned_parameters = ast.literal_eval(banned_parameters)  # Convert string to list
    
        if isinstance(not_banned_parameters, str):
            not_banned_parameters = ast.literal_eval(not_banned_parameters)  # Convert string to list
    else:
        banned_parameters = []
        not_banned_parameters = []
    
    # Ensure they are lists
    banned_parameters = banned_parameters if isinstance(banned_parameters, list) else []
    not_banned_parameters = not_banned_parameters if isinstance(not_banned_parameters, list) else []

    print(f"Banned Parameters for {sheet_name}: {banned_parameters}")
    print(f"Not Banned Parameters for {sheet_name}: {not_banned_parameters}")

    # First 16 columns remain unchanged
    first_16_columns = df.columns[:16].tolist()

    # Function to get associated columns for a parameter
    def get_param_columns(param):
        return [col for col in [param, f"{param} Compliance", f"FSSR Limits {param}"] if col in df.columns]

    # Collect reordered columns based on banned and not banned parameters
    ordered_columns = first_16_columns  # Start with the first 16 columns
    
    # Add banned parameters first
    bn=[]

    for param in banned_parameters:
        ordered_columns.extend(get_param_columns(param))
        bn.extend(get_param_columns(param))
        print(param)
        print("\n")
        print("ye hi hai")
        print("\n")
        # print("done")

    # Insert the "Separator" column if there are banned parameters
   
    n_bn=[]
    if banned_parameters and not_banned_parameters:
        ordered_columns.append("Separator")
        # print(commodity_row["Commodity"])
        # print("separator added")


    # Add not banned parameters next
    for param in not_banned_parameters:
        print(param)
        print("\n")
        n_bn.extend(get_param_columns(param))
        # print(get_param_columns(param))
        ordered_columns.extend(get_param_columns(param))

    # Identify any remaining columns that are not part of the first 16 or the ordered parameters
    remaining_columns = [col for col in df.columns if col not in ordered_columns]

    # Append the remaining columns to preserve all data
    ordered_columns.extend(remaining_columns)

    # Keep only the columns that exist in the original sheet, plus the new "Separator" column
    existing_columns = [col for col in ordered_columns if col in df.columns or col == "Separator"]

    # Reorder the DataFrame based on the existing columns
    df = df.reindex(columns=existing_columns)

    # print(bn)
    # print(n_bn)

    # Store cleaned sheet
    sheet_dict[sheet_name] = df

# Save the updated Excel file
with pd.ExcelWriter(output_file, engine="openpyxl") as writer:
    for sheet_name, sheet_df in sheet_dict.items():
        sheet_df.to_excel(writer, sheet_name=sheet_name, index=False)

print("✅ Successfully reordered parameters and added a 'Separator' column.")
```
